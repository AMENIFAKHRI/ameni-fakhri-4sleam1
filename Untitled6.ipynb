```python
import librosa
import numpy as np

def extract_spectrogram(file_path):
    try:
        audio, sr = librosa.load(file_path, sr=16000)

        audio = audio[:16000]  # 1 seconde

        spec = librosa.feature.melspectrogram(y=audio, sr=sr)
        spec_db = librosa.power_to_db(spec, ref=np.max)

        return spec_db

    except:
        return None
```

```python
df_final = pd.concat([df1, df_not_esc], ignore_index=True)

print(df_final["is_cry"].value_counts())
```

```python
df_cry = df_final[df_final.is_cry == 1]
df_not = df_final[df_final.is_cry == 0]

min_size = min(len(df_cry), len(df_not))

df_cry = df_cry.sample(min_size, random_state=42)
df_not = df_not.sample(min_size, random_state=42)

df_balanced = pd.concat([df_cry, df_not])
df_balanced = df_balanced.sample(frac=1).reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(
    df_balanced, test_size=0.2, random_state=42
)
```

```python
def extract_spectrogram(file_path):
    try:
        # 🔥 ignorer ogg
        if not file_path.endswith(".wav"):
            return None

        audio, sr = librosa.load(file_path, sr=16000)

        if audio is None or len(audio) < 1000:
            return None

        audio = audio[:16000]

        spec = librosa.feature.melspectrogram(y=audio, sr=sr)
        spec_db = librosa.power_to_db(spec, ref=1.0)

        return spec_db

    except:
        return None
```

```python
def build_dataset(df):
    X, y = [], []

    for _, row in df.iterrows():
        spec = extract_spectrogram(row["filename"])

        if spec is not None:
            X.append(spec)
            y.append(row["is_cry"])

    return np.array(X), np.array(y)

X_train, y_train = build_dataset(train_df)
X_test, y_test = build_dataset(test_df)
```

```python
X_train = np.nan_to_num(X_train)
X_test = np.nan_to_num(X_test)

X_train = X_train / (np.max(X_train) + 1e-6)
X_test = X_test / (np.max(X_test) + 1e-6)

X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Input(shape=X_train.shape[1:]),

    Conv2D(16, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(32, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),

    Dense(64, activation='relu'),
    Dropout(0.3),

    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train, y_train,
    epochs=12,
    batch_size=16,
    validation_data=(X_test, y_test)
)
```

```python
from flask import Flask, request, jsonify
import librosa
import numpy as np
from tensorflow.keras.models import load_model

app = Flask(__name__)

model = load_model("cnn_model.h5")

def process(file):
    audio, sr = librosa.load(file, sr=16000)
    audio = audio[:16000]

    spec = librosa.feature.melspectrogram(y=audio, sr=sr)
    spec = librosa.power_to_db(spec, ref=1.0)

    spec = np.nan_to_num(spec)
    spec = spec / (np.max(spec) + 1e-6)

    spec = spec[..., np.newaxis]
    spec = np.expand_dims(spec, axis=0)

    return spec

@app.route("/predict", methods=["POST"])
def predict():
    file = request.files["file"]
    file.save("temp.wav")

    data = process("temp.wav")

    pred = model.predict(data)[0][0]

    result = "cry" if pred > 0.5 else "not_cry"

    return jsonify({
        "result": result,
        "confidence": float(pred)
    })

app.run(host="0.0.0.0", port=5000)
```
